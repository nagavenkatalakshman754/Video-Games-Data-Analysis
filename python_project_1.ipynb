```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
video_games = pd.read_csv("video_games.csv")

# 1. Get sample of 5 rows
print("Sample of 5 rows:")
print(video_games.sample(5))
```

```python
# 2. Display all the columns
print("Columns in the dataset:")
print(video_games.columns)
```

```python
# 3. Display the dimensions
print("Dimensions of the dataset:")
print(video_games.shape)
```

```python
# 4. Display all columns data types
print("Data types of each column:")
print(video_games.dtypes)
```

```python
# 5. Display Top 10 rows
print("Top 10 rows:")
print(video_games.head(10))
```

```python
# 6. Bottom 10 rows
print("Bottom 10 rows:")
print(video_games.tail(10))
```

```python
# 7. Display all Racing Games which have Critic_score >= 95.  
#Assuming 'Metrics.Review Score' is the critic score,as critic score is not present
print("Racing Games with Critic Score >= 95:")
racing_games = video_games[video_games['Metadata.Genres'].str.contains('Racing', na=False)]
print(racing_games[racing_games['Metrics.Review Score'] >= 95])
```

```python
# 8. Video games with User_score > 9 (assuming Review Score, as User_score isn't in the provided columns)
print("Video games with Review Score > 9:")
high_review_games = video_games[video_games['Metrics.Review Score'] > 9]
print(high_review_games)
```

```python
# 9. Display all unique user_scores  (here there is no user_scores so i took review scores)
print("Unique Review Scores:")
print(video_games['Metrics.Review Score'].unique())
```

```python
# 10. Display all types of Games (a game type is defined by a genre & type of genre combination)
print("All game genres:")
print(video_games['Metadata.Genres'].unique())
```

```python
# 11. Access row 15
print("Row 15:")
print(video_games.iloc[14])
```

```python
# 12. Display EU_Sales, Global_Sales and Critic_Count of all games using iloc
# There are no EU_Sales or Critic_Count in the provided column names. I have to assume 'Metrics.Sales' is Global Sales
print("Global Sales using iloc:")
print(video_games.iloc[:, 16])
```

```python
# 13. Display the Name, Genre, and Developer of all games using loc
#There is no 'Developer' column. Assuming the publisher as developer here TQ.
print("Name, Genre, and Publisher of all games using loc:")
print(video_games.loc[:, ['Title', 'Metadata.Genres', 'Metadata.Publishers']])
```

```python
# 14. Change user_data there is no such column so i take review score data type to float and display data types
video_games['Metrics.Review Score'] = pd.to_numeric(video_games['Metrics.Review Score'], errors='coerce')
print("Data Types after changing Review Score to float:")
print(video_games.dtypes)
```

```python
# 15. Which publisher has made how much sales in USD globally?
# I have to assume 'Metrics.Sales' is Global Sales
print("Global Sales by Publisher:")
global_sales_by_publisher = video_games.groupby('Metadata.Publishers')['Metrics.Sales'].sum()
print(global_sales_by_publisher)
```

```python
# 16. Remove all null values from User_Score
# there is no user_score so i hve to assume review.score
video_games['Metrics.Review Score'] = video_games['Metrics.Review Score'].fillna(0)
print("Review Scores after filling Null values:")
print(video_games['Metrics.Review Score'])
```

```python
# 17. Plot the histogram chart with x=User_Score and y=Publisher (Placeholder)
# Since there is no User_Score, I'll plot a histogram of 'Metrics.Sales'
plt.hist(video_games['Metrics.Sales'], bins=10)
plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Global Sales')
plt.show()
```

```python
# 18. Display all video games.
print("All video games:")
print(video_games)
```

```python
# 19. Update Metrics.Sales with + 5: (map with lambda)
video_games['Metrics.Sales'] = video_games['Metrics.Sales'].map(lambda x: x + 5 if pd.notna(x) else x)
print("Global Sales after adding 5:")
print(video_games['Metrics.Sales'])
plt.hist(video_games['Metrics.Sales'], bins=10, alpha=0.7)
plt.xlabel('Updated Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Updated Sales')
plt.show()
```

```python
# 20. Display which Genre has how many games
all_games = video_games.groupby('Metadata.Genres')['Metadata.Genres'].count()
print("Number of games by Genre:")
print(all_games)
```

```python
# 21. Create a bar chart for all games.
plt.figure(figsize=(10,6))
plt.bar(all_games.index, all_games.values)
plt.xlabel('Genre')
plt.ylabel('Number of games')
plt.title('Number of Games by Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 22. Which publisher has most games sold: top 20 publishers.
top_publishers = video_games.groupby('Metadata.Publishers')['Metrics.Sales'].sum().sort_values(ascending=False).head(20)
print("Top 20 Publishers by Global Sales:")
print(top_publishers)
```

```python
# 23. Plot the bar chart for the above top 20 publishers.
plt.figure(figsize=(12,8))
plt.bar(top_publishers.index, top_publishers.values)
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Global Sales by Top 20 Publishers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
